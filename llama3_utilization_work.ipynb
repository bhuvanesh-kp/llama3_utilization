```python
import torch
import json
from transformers import AutoTokenizer,AutoModelForCausalLM,BitsAndBytesConfig,pipeline
```

```python
token = json.load(open("config.json"))
```

```python
HG_token = token['Hugging_face_token']
```

```python
model = 'meta-llama/Meta-Llama-3-8B'
```

**Quantization for utilizing the llama3 8B model with in the limitations of google colab**

```python
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type='nf4',
    bnb_4bit_compute_dtype=torch.bfloat16
)
```

*Obtain model tokens*

```python
tokenizer = AutoTokenizer.from_pretrained(model,token=HG_token)

tokenizer.pad_token = tokenizer.eos_token
```

```python
Model = AutoModelForCausalLM.from_pretrained(model,
                                             device_map='auto',
                                             quantization_config=bnb_config,
                                             token=HG_token)
```

```python
text_generation = pipeline("text-generation",
                           model=Model,
                           tokenizer = tokenizer,
                           max_new_tokens = 128)
```

```python
prompt = "Who is Mark Zukerburg in Meta?"
```

```python
llama3_responce = text_generation(prompt)
llama3_responce
```

```python
llama3_responce_1 = text_generation(prompt)
llama3_responce_1
```

```python
def get_response(text):
  responce = text_generation(text)
  llama_responce = responce[0]['generated_text']
  return llama_responce
```

```python
new_prompt = "How is the founder of OpenAI"


llama_responce_2 = get_response(new_prompt)
print(llama_responce_2)
```
