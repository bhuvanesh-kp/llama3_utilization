```python
from langchain.document_loaders import TextLoader
```

```python
loader = TextLoader("/content/sample.txt")
```

```python
document = loader.load()
```

```python
print(document[0].metadata)
```

```python
print(f"Your document has {len(document[0].page_content)} characters")
```

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=400
)
```

```python
document_chunks = text_splitter.split_documents(document)
```

```python
print(f"no.of chunks in the document is {len(document_chunks)}")
```

```python
from langchain.vectorstores import FAISS
```

```python
!pip install faiss-gpu
```

```python
stored_embeddings = FAISS.from_documents(document_chunks, embeddings)
```

```python
from langchain.chains import RetrievalQA
```

```python
QA_chain = RetrievalQA.from_chain_type(
    llm=hf,
    chain_type="stuff",
    retriever=stored_embeddings.as_retriever()
)
```

```python
question = """
in which year transfomer came to play a important role in LLM
"""
```

```python
responce = QA_chain.run(question)
```
